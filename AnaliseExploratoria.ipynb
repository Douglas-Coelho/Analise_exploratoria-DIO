```python
#Importanto biblioteca Pandas e habilitando a edição de gráficos e aplicando um estilo

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Outra maneira de fazer upload dos arquivos, esse é via código
from google.colab import files
arq = files.upload()
```

```python
#Lendo o arquivo do Excel
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Mostrando as 5 primeiras linhas
df.head()
```

```python
#Identificado os tipos

df.dtypes
```

```python
#Descobrindo a receita total

df["Valor Venda"].sum()
```

```python
#Criando a coluna Custo

df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
#Qual o custo total?
#Vou usar "round" para conseguir mexer nas casas decimais

round(df["Custo"].sum(), 2)
```

```python
#Achando o valor do lucro

df["Lucro por unidade"] = df["Preço Unitário"] - df["Custo Unitário"]
```

```python
df.head(1)
```

```python
#Lucro total da venda

df["Lucro da Venda"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(4)
```

```python
#Sabendo o lucro total

round(df["Lucro da Venda"].sum(), 2)
```

```python
#Coluna com total de dias do envio do produto

df["Dias Ate o Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(3)
```

```python
df["Dias Ate o Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
df["Dias Ate o Envio"].dtype
```

```python
#Média de dias da entrega

round(df.groupby("Marca")["Dias Ate o Envio"].mean(), 2)
```

```python
#Verificando se temos valores ausentes

df.isnull().sum()
```

```python
#E SE QUISERMOS SABER O LUCRO POR ANO E POR MARCA?
```

```python
df["Lucro da Venda"].dtype
```

```python
#Agrupando por ano e marca

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro da Venda"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando Index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro da Venda"].sum().reset_index()
lucro_ano
```

```python
#Achando o total de produtos vendidos

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).reset_index()
```

```python
#Gráfico total de produtos vendidos

df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos Vendidos")
plt.xlabel("Produto")
plt.ylabel("Quantidade Vendida");
```

```python
#Extraindo o lucro por ano

df.groupby(df["Data Venda"].dt.year)["Lucro da Venda"].sum().plot.bar(title="Ano x Venda")
plt.xlabel("Ano da Venda")
plt.ylabel("Valor Vendido");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro da Venda"].sum().reset_index()
```

```python
#Selecionando apenas as vendas de 2009

vendas_2009 = df[df["Data Venda"].dt.year == 2009]
vendas_2009.head()
```

```python
#Filtrando as vendas do mês de 2009

vendas_2009.groupby(vendas_2009["Data Venda"].dt.month)["Lucro da Venda"].sum().plot(title="Lucro x Marca")
plt.xlabel("Mês")
plt.ylabel("Valor Vendido");
```

```python
vendas_2009.groupby("Marca")["Lucro da Venda"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
vendas_2009.groupby("Classe")["Lucro da Venda"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
#Análise estatística do nosso "Tempo de envio"
#Legenda:
#count = número de itens da nossa coluna
#mean = Média da nossa coluna
#std = Desvio Padrão da nossa coluna
#min = Mínimo da nossa coluna
#25% = Data em que 25% dos envios foram feitos e o mesmo vale para 50 e 75, nas suas respectivas proporções
#max = Tempo máximo do envio

df["Dias Ate o Envio"].describe()
```

```python
'''Fazendo análise gráfica de um "box", onde a linha inferior representa o valor mínimo, a primeira parte da caixa é o "Q1", que representa os 25% dos dados, a linha verde representa o "Q3" que é a mediana, a parte superior do box
representa o valor que corresponde a 75% dos dados e a bolinha na parte superior representa a disparidade que há no gráfico, representando o valor máximo também chamado de outlier(Um dado que se destaca dos demais'''
plt.boxplot(df["Dias Ate o Envio"]);
```
